NECESSARY PACKAGES & DATA LOADING

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf    
```

```python
#load data
log_returns = pd.read_csv('DATA/ETF_log_returns.csv', index_col='Date', parse_dates=True)
close_returns = pd.read_csv('DATA/ETF_ADJ_close_returns.csv', index_col='Date', parse_dates=True)
# Check it loaded:
# log_returns
# close_returns
```

STRATEGY FORMATION

```python
etfs = ['DIA','SPY','QQQ']

#creating rolling averages
short_window = 50
long_window = 200

short_ma = close_returns.rolling(short_window).mean()
long_ma = close_returns.rolling(long_window).mean()
short_ma = pd.DataFrame(short_ma, index=close_returns.index, columns=close_returns.columns)
short_ma
long_ma = pd.DataFrame(long_ma, index=close_returns.index, columns=close_returns.columns)
long_ma
```

VISUALIZATION

```python
def plot_etf_ma(etf_list, close_returns, short_ma, long_ma):
    for etf in etf_list:
        plt.rcParams['figure.figsize'] = 20,10
        plt.grid(True, alpha=0.3)
        
        plt.plot(close_returns[etf], label=f'{etf} Returns')
        plt.plot(short_ma[etf], label='50-day MA')
        plt.plot(long_ma[etf], label='200-day MA')
        
        plt.legend(loc=2)
        plt.title(f'{etf} Returns with Moving Averages')
        plt.show()
plot_etf_ma(etfs, close_returns, short_ma, long_ma)
```

```python
def ma_crossover_strategy(etfs, close_returns, log_returns, short_ma, long_ma):
    log_return_summary = {}  # To store total log return sums for each ETF defined at end
    etf_dfs = {} #New dictionary to store full dataframes per ETF

    for etf in etfs: #loops it for each etf in etf list ('SPY','DIA,'QQQ')
        # Build DataFrame
        df = pd.DataFrame({
            'close_ret': close_returns[etf],
            'log_ret': log_returns[etf],
            'short_ma': short_ma[etf],
            'long_ma': long_ma[etf]
        }).dropna()

        # Entry and exit signals
        df['entry_signal'] = (df['short_ma'] > df['long_ma']) & (df['short_ma'].shift(1) <= df['long_ma'].shift(1))
        df['exit_signal']  = (df['short_ma'] < df['long_ma']) & (df['short_ma'].shift(1) >= df['long_ma'].shift(1))
        #meaning that the entry signal has have the short ma greater than the long ma and the short ma has to have been smaller thant he long ma yesterday(shift(1))
        #also which means the it must have crossed over with todays close in order to enter the trade. 
        #this assigns Boolean values (True or False) so row values will be either True or False

        # Initialize position
        df['position'] = np.nan #assigning NAN values to empty column [position]
        
        # Set entry points
        df.loc[df['entry_signal'], 'position'] = 1
        #where entry_signal is True it assigns 1 to position value
        #loc[] acts as a filter; only the rows where the condition is True are selected
        # Set exit points
        df.loc[df['exit_signal'],  'position'] = 0
        #where exit signal is True it assigns 0 to positon value
        
        # Forward-fill the position
        df['position'] = df['position'].ffill().fillna(0)
        #ffill() filles foward any NaN(missing values) with the last known non-missing value above it 
        #until it reaches the next value and now does it with this value and so on 
        #fill na (0) ensures 
        #out of market at the start if no signal has occurred

        # Shift to avoid lookahead bias
        df['shifted_position'] = df['position'].shift(1) 
# We shift the 'position' by 1 to create 'shifted_position',
# because we can't know today's closing price (and thus the crossover signal)
# until after the market closes. This means we can't act on today's signal today.
# So, the actual trade must occur the next day.
# By using position from t-1, we ensure the strategy avoids lookahead bias
# and only earns returns when we were already in the trade.
 

        # Calculate strategy returns
        df['strategy_log_return'] = df['shifted_position'] * df['log_ret']
        #simply the psotion multiplied by the % return (log) which is additive!.
        #if in a position (1) we multiply return cause we are earning that
        #if not in a position(0) we multply it also by the return but will get 0 because we are not earning the return

        # Drop NA rows the shift will cause a us to lose n-1 data again due to looking at yesterdays value 
        df.dropna(inplace=True)
        

        # Wealth index
        df['strategy_wealth'] = np.exp(df['strategy_log_return'].cumsum())
        df['buy_hold_wealth'] = np.exp(df['log_ret'].cumsum())
        #this caclulates the running total(sum) of both strategy and it works because log returns are additive 
        #and then we back transform to get it back into dollar value.
        
        #they will both not start at 1 since first day is the first days return actually but can be implemented.
    

        # Plot
        plt.figure(figsize=(14, 6))
        plt.plot(df['strategy_wealth'], label='Strategy ($1 invested)')
        plt.plot(df['buy_hold_wealth'], label='Buy & Hold ($1 invested)')
        plt.title(f'{etf} – Correct MA Crossover Strategy vs Buy & Hold')
        plt.ylabel('Wealth Index')
        plt.xlabel('Date')
        plt.legend()
        plt.grid(alpha=0.3)
        plt.show()
       
        etf_dfs[etf] = df.copy()

    # Store sum of strategy and buy-hold log returns
        log_return_summary[etf] = {
        'Total Strategy Return (%)': round((np.exp(df['strategy_log_return'].sum()) - 1) * 100, 2),
        'Total Buy & Hold Return (%)': round((np.exp(df['log_ret'].sum()) - 1) * 100, 2)
}

    return pd.DataFrame(log_return_summary).T , etf_dfs


       

summary, dfs = ma_crossover_strategy(etfs, close_returns, log_returns, short_ma, long_ma) #assigning multiple variables at once to return of the function (goes based on return order)
# Display results table


```

```python
#calculating anual standard deviation for strategy 
for etf in dfs:
    daily_std = dfs[etf]['strategy_log_return'].std()
    annualized_std = daily_std * (252 ** 0.5)

    print(f"{etf} — Annualized Std Dev (Volatility): {round(annualized_std, 6)}")
```

```python
#Calculating annualized standard deviation for buy and hold
for etf in dfs:
    daily_std = dfs[etf]['log_ret'].std()
    annualized_std = daily_std * (252 ** 0.5)

    print(f"{etf} — Annualized Std Dev (Volatility): {round(annualized_std, 6)}")
```

```python
#rolling standard deviation using 21 day rolling window on strategy

rolling_window = 21  # 1-month rolling window

for etf in dfs:
    df = dfs[etf]

    # Calculate rolling std dev (annualized)
    df['rolling_std'] = df['strategy_log_return'].rolling(window=rolling_window).std() * (252 ** 0.5)

    # Plot full history of rolling std
    plt.figure(figsize=(14, 5))
    plt.plot(df['rolling_std'], label=f'{etf} Rolling Volatility')
    plt.title(f'{etf} – Rolling Annualized Standard Deviation (Volatility)\nWindow = {rolling_window} days')
    plt.xlabel('Date')
    plt.ylabel('Annualized Std Dev')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()
```

```python
#rolling standard deviation using 21 day rolling window on buy and hold 

rolling_window = 21  # 1-month rolling window

for etf in dfs:
    df = dfs[etf]

    # Calculate rolling std dev (annualized)
    df['rolling_std'] = df['log_ret'].rolling(window=rolling_window).std() * (252 ** 0.5)

    # Plot full history of rolling std
    plt.figure(figsize=(14, 5))
    plt.plot(df['rolling_std'], label=f'{etf} Rolling Volatility')
    plt.title(f'{etf} – Rolling Annualized Standard Deviation (Volatility)\nWindow = {rolling_window} days')
    plt.xlabel('Date')
    plt.ylabel('Annualized Std Dev')
    plt.grid(alpha=0.3)
    plt.legend()
    plt.show()
```

```python
#annualized sharpe ratio for each using a rf = 0 for strategy 
for etf in dfs:
    daily_returns = dfs[etf]['strategy_log_return']

    mean_daily = daily_returns.mean()
    std_daily = daily_returns.std()

    annual_return = mean_daily * 252
    annual_std = std_daily * (252 ** 0.5)

    sharpe_ratio = annual_return / annual_std

    print(f"{etf} — Sharpe Ratio: {round(sharpe_ratio, 4)}")
```

```python
#annualized sharpe ratio for each using a rf = 0 for buy and hold 
for etf in dfs:
    daily_returns = dfs[etf]['log_ret']

    mean_daily = daily_returns.mean()
    std_daily = daily_returns.std()

    annual_return = mean_daily * 252
    annual_std = std_daily * (252 ** 0.5)

    sharpe_ratio = annual_return / annual_std

    print(f"{etf} — Sharpe Ratio: {round(sharpe_ratio, 4)}")
```

```python
# Store Sharpe ratios in a summary table
sharpe_summary = []

for etf in dfs:
    strat_daily = dfs[etf]['strategy_log_return']
    hold_daily = dfs[etf]['log_ret']

    strat_sharpe = (strat_daily.mean() * 252) / (strat_daily.std() * (252 ** 0.5))
    hold_sharpe = (hold_daily.mean() * 252) / (hold_daily.std() * (252 ** 0.5))

    sharpe_summary.append({
        'ETF': etf,
        'Strategy Sharpe': round(strat_sharpe, 4),
        'Buy & Hold Sharpe': round(hold_sharpe, 4)
    })

# Create a DataFrame
sharpe_df = pd.DataFrame(sharpe_summary)
print(sharpe_df)
```

ADDING A .01% TRESHOLD DIFFERENCE 

```python

def ma_crossover_strategy_treshold(etfs, close_returns, log_returns, short_ma, long_ma):
    log_return_summary = {}  
    etf_dfs = {} 

    for etf in etfs: 
        df = pd.DataFrame({
            'close_ret': close_returns[etf],
            'log_ret': log_returns[etf],
            'short_ma': short_ma[etf],
            'long_ma': long_ma[etf]
        }).dropna()

        # Calculate the % difference between short and long MA
        df['ma_diff_pct'] = ((df['short_ma'] - df['long_ma']) / df['long_ma']) * 100

        # Add 0.01% threshold filter to entry and exit signals
        df['entry_signal'] = (
        (df['short_ma'] > df['long_ma']) &
        (df['short_ma'].shift(1) <= df['long_ma'].shift(1)) &
        (df['ma_diff_pct'] > 0.01)
        )

        df['exit_signal'] = (
        (df['short_ma'] < df['long_ma']) &
        (df['short_ma'].shift(1) >= df['long_ma'].shift(1)) &
        (df['ma_diff_pct'] < -0.01)
        )
   
        # Initialize position
        df['position'] = np.nan
        
        # Set entry points
        df.loc[df['entry_signal'], 'position'] = 1
        df.loc[df['exit_signal'],  'position'] = 0
            
        # Forward-fill the position
        df['position'] = df['position'].ffill().fillna(0)

        # Shift to avoid lookahead bias
        df['shifted_position'] = df['position'].shift(1) 

        # Calculate strategy returns
        df['strategy_log_return'] = df['shifted_position'] * df['log_ret']

        # Drop NA rows the shift will cause a us to lose n-1 data again due to looking at yesterdays value 
        df.dropna(inplace=True)    

        # Wealth index
        df['strategy_wealth'] = np.exp(df['strategy_log_return'].cumsum())
        df['buy_hold_wealth'] = np.exp(df['log_ret'].cumsum())
    
        # Plot
        plt.figure(figsize=(14, 6))
        plt.plot(df['strategy_wealth'], label='Strategy ($1 invested)')
        plt.plot(df['buy_hold_wealth'], label='Buy & Hold ($1 invested)')
        plt.title(f'{etf} – Correct MA Crossover Strategy vs Buy & Hold')
        plt.ylabel('Wealth Index')
        plt.xlabel('Date')
        plt.legend()
        plt.grid(alpha=0.3)
        plt.show()

        etf_dfs[etf] = df.copy()

    # Store sum of strategy and buy-hold log returns
        log_return_summary[etf] = {
        'Total Strategy Return (%)': round((np.exp(df['strategy_log_return'].sum()) - 1) * 100, 2),
        'Total Buy & Hold Return (%)': round((np.exp(df['log_ret'].sum()) - 1) * 100, 2)
}

    return pd.DataFrame(log_return_summary).T, etf_dfs


       

summary_2, dfs_2 = ma_crossover_strategy_treshold(etfs, close_returns, log_returns, short_ma, long_ma)
# Display results table


```
