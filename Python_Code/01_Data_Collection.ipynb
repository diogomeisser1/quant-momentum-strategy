```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf    
```

DATA COLLECTION

```python
etfs = ['QQQ','SPY','DIA'] #QQQ-> NASDAQ, DIA -> Dow Jones
close_returns = yf.download(tickers=etfs, interval= '1d')['Close'] #adjusted close = TRUE
close_returns
```

```python
close_returns.dropna(inplace=True) #aligning Index
close_returns
```

```python
log_returns = np.log(close_returns[etfs] / close_returns[etfs].shift(1)).dropna() #calculating returns
```

EXPORTING DATA

```python
log_returns.to_csv('ETF_log_returns.csv')
close_returns.to_csv('ETF_ADJ_close_returns.csv')
```
